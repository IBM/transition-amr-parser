```python
import matplotlib.pyplot as plt
```

### base model (6x6), on AMR 2.0

```python
beams = list(range(1, 11))
scores = [81.23, 81.46, 81.67, 81.71, 81.69, 81.67, 81.76, 81.76, 81.77, 81.78]
scores = list(map(lambda x: round(x, 2), scores))
```

```python
scores_avg3 = [81.14, 81.34, 81.50, 81.53, 81.54, 81.57, 81.65, 81.64, 81.70, 81.69]
scores_avg3 = list(map(lambda x: round(x, 2), scores_avg3))
```

```python
scores_best1 = [81.04, 81.23, 81.47, 81.48, 81.49, 81.51, 81.50, 81.53, 81.55, 81.56]
scores_best1 = list(map(lambda x: round(x, 2), scores_best1))
```

```python
scores_last = [80.83, 81.15, 81.28, 81.31, 81.31, 81.34, 81.36, 81.35, 81.33, 81.37]
scores_last = list(map(lambda x: round(x, 2), scores_last))
```

```python
plt.figure(figsize=(4, 3))
plt.plot(beams, scores, 'r-x')
plt.xlabel('beam size', fontsize=7.5)
plt.ylabel('Smatch (%)', fontsize=7.5)
plt.ylim([81.1, 81.9])
# plt.xlim([1, 10])
plt.grid()
# plt.savefig('beam_smatch_2digits.pdf', bbox_inches='tight', dpi=200)
plt.show()
```

```python
scores_round = list(map(lambda x: round(x, 1), scores))
scores_avg3_round = list(map(lambda x: round(x, 1), scores_avg3))
scores_best1_round = list(map(lambda x: round(x, 1), scores_best1))
scores_last_round = list(map(lambda x: round(x, 1), scores_last))
```

```python
plt.figure(figsize=(4, 3))
plt.plot(beams, scores_round, 'r-x')
plt.xlabel('beam size', fontsize=7.5)
plt.ylabel('Smatch (%)', fontsize=7.5)
plt.ylim([81.1, 81.9])
plt.grid()
# plt.savefig('beam_smatch.pdf', bbox_inches='tight', dpi=200)
plt.show()
```

```python
plt.figure(figsize=(4, 3))
plt.plot(beams, scores_round, 'r-x')
plt.plot(beams, scores_avg3_round, 'b-v')
plt.xlabel('beam size', fontsize=7.5)
plt.ylabel('Smatch (%)', fontsize=7.5)
plt.ylim([81.0, 81.9])
plt.legend(['avg best5', 'avg best3'])
plt.grid()
# plt.savefig('beam_smatch_avg53.pdf', bbox_inches='tight', dpi=200)
plt.show()
```

```python
plt.figure(figsize=(4, 3))
plt.plot(beams, scores_round, 'r-x')
plt.plot(beams, scores_avg3_round, 'b-v')
plt.plot(beams, scores_best1_round, 'c-^')
plt.xlabel('beam size', fontsize=7.5)
plt.ylabel('Smatch (%)', fontsize=7.5)
plt.ylim([80.9, 81.9])
plt.legend(['avg best5', 'avg best3', 'best1'])
plt.grid()
# plt.savefig('beam_smatch_avg531.pdf', bbox_inches='tight', dpi=200)
plt.show()
```

```python
plt.figure(figsize=(4, 3))
plt.plot(beams, scores_round, 'r-x')
plt.plot(beams, scores_avg3_round, 'b-v')
plt.plot(beams, scores_best1_round, 'c-^')
plt.plot(beams, scores_last_round, 'g-o')
plt.xlabel('beam size', fontsize=7.5)
plt.ylabel('Smatch (%)', fontsize=7.5)
plt.ylim([80.7, 81.9])
plt.legend(['avg best5', 'avg best3', 'best1', 'last'])
plt.grid()
# plt.savefig('beam_smatch_avg531_last.pdf', bbox_inches='tight', dpi=200)
plt.show()
```
